```python
# read LIGO-LW .xml files
psd_xml = "../data/psd/H1L1V1-REF_PSD-1257382107-1257388107.xml"
snr_xml = "../data/coinc/snr/H1L1V1_1187006031_3_432.xml"
coinc_xml = "../data/coinc/H1L1V1_1187008603_3_606.xml"
# snr_xml = "../data/coinc/snr/H1L1V1_1187006031_3_432_no_ilwdchar.xml"
# coinc_xml = "../data/coinc/H1L1V1_1187008603_3_606_no_ilwdchar.xml"
mock_xml = "../data/coinc/mock.xml"

# arrays to get by instrument name
ifos = ("H1", "L1", "V1")

files = [coinc_xml for _ in range(20)]
```

```python
df = spiir.io.ligolw.load_ligolw_tables(
    paths=[files[0]],
    table="postcoh",
    ilwdchar_compat=True,
    legacy_postcoh_compat=True,
    verbose=True,
    nproc=1,
)
df.T
```

```python
psds = spiir.io.ligolw.load_all_ligolw_frequency_arrays(files[0])
psds = pd.DataFrame(psds)
psds
```

```python
snr_series = spiir.io.ligolw.load_all_ligolw_snr_arrays(snr_xml, add_epoch_time=False, ilwdchar_compat=True)
df = pd.DataFrame(snr_series.values()).T.fillna(np.nan)
df
```

```python
df.apply(lambda snr: np.sqrt(np.real(snr)**2 + np.imag(snr)**2), axis=0).plot(title="SNR Series", figsize=(12,8))
```
